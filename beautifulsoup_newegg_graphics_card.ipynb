```python
import requests
from bs4 import BeautifulSoup as bs
import time
import re
import pandas as pd
```

```python
r = requests.get('https://www.newegg.com/p/pl?d=graphics+card')
webpage = bs(r.content)
```

```python
#get total number of pages
buttons = webpage.find_all('button', {'class':'btn'})
buttons_num = []
for button in buttons:
    try:
        page = int(button.text)
        buttons_num.append(page)
    except:
        continue
        
num_pages = buttons_num[-1]
```

```python
product_names = []
product_prices = []
product_shippings = []
for page in range(1,num_pages+1):
    r = requests.get('https://www.newegg.com/p/pl?d=graphics+card&page={page}')
    time.sleep(3) #wait 5 sec for each request
    webpage = bs(r.content)
    

    all_products = webpage.find_all('div', {'class':'item-cells-wrap border-cells items-grid-view four-cells expulsion-one-cell'})
    
    
    for product in all_products:
        price_int = product.find('li', {'class':'price-current'}).strong.text
        price_int = float(re.sub(',', '', price_int))
        price_decimal = float(product.find('li', {'class':'price-current'}).sup.text)
        product_prices.append(price_int + price_decimal)
        product_names.append(product.find('a', {'class':'item-title'}).text)
        product_shippings.append(product.find('li', {'class':'price-ship'}).text)
        
        
        
    print("We've looked at up to page:", page)
```

```python
item_info = dict()
item_info['name'] = product_names
item_info['price'] = product_prices
item_info['shipping'] = product_shippings
```

```python
#make dataframe of graphics cards
graphics_cards = pd.DataFrame(item_info)
```
